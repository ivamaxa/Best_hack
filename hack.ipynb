Preprocessing dataset. Check impact on the data on every step of preprocessing data. Using cross-validation on different ML models to do that.

Подготовка данных. Смотрим как влияют наши действия на данные при помощи кросс-валидации на разных моделях.

```python
# heuristic: according to description following features won't impact model accuracy. Thus drop 'em.
# эвристика: согласно описанию, следующие признаки не повлияют на точность модели. Поэтому избавимся от них.
df1 = df.copy().drop(columns=['Shrt_Desc', 'GmWt_Desc1', 'GmWt_Desc2', 'GmWt_1', 'GmWt_2'])
```

```python
# import required libraries and set up cross-val parameters
# импортируем необходимые библиотеки и устанавливаем параметры для кросс-валидации
from sklearn.model_selection import KFold
cv = KFold(5, True, 0)
from sklearn.model_selection import cross_val_score
```

```python
# importing required models and defining 'em. The following choice of models explained by different models 'nature'.
# Импортируем и определяем необходимые модели. Выбор следующих моделей объясняется их разной 'природой'.
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()

from sklearn.linear_model import LinearRegression
linres = LinearRegression()

from sklearn.linear_model import Ridge
ridge = Ridge()

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=0) # define n_estimators to avoid warning system messages

from sklearn.svm import SVR
svm = SVR(gamma='scale') # define gamma to avoid warning system messages

from sklearn.neighbors import KNeighborsRegressor
k_near = KNeighborsRegressor()

# making lists of labels and models respectivly
# создаем массив ярлыков и моделей соответственно
labels = ['DecisionTree', 'LinRess', 'Ridge', 'RandomForest', 'K_neighbours', 'SVM']
clf = [dtr, linres, ridge, rf, k_near, svm]
```

Get rid of missing values is crucial part of preprocessing. There are two ways to do so without loosing valuable data: replace with mean or replace with median. Let's test both the approaches and see which one works best.

Избавиться от пропущенных значений- важный шаг в подготовке данных. Существует 2 способа избавиться от пропущенных значений, чтобы не потерять важные данные: заменить средним или заменить медианной. Давайте протестируем оба способа и посмотрим какой работает лучше.

```python
# replace missing values with mean
# заменяем пропущенный значения средним
df_mean = df1.copy().fillna(df1.mean())
X_mean = df_mean.drop(columns=['Energ_Kcal'])
y_mean = df_mean['Energ_Kcal']

score_mean = []
for model in clf:
    score_mean.append(-1*cross_val_score(model, X_mean, y_mean, cv=cv, scoring='neg_mean_absolute_error').mean())
print(score_mean)
sns.barplot(x=labels, y=score_mean)
```

```python
# replace missing values with median
# заменяем пропущенный значения медианной
df_median = df1.copy().fillna(df1.median())
X_median = df_median.drop(columns=['Energ_Kcal'])
y_median = df_median['Energ_Kcal']

score_median = []
for model in clf:
    score_median.append(-1*cross_val_score(model, X_median, y_median, cv=cv, scoring='neg_mean_absolute_error').mean())
print(score_median)
sns.barplot(x=labels, y=score_median)
```

Based on tests, we choose median as a replacement for missing values, due to slightly better result on cross-validation. Next step: ouliers removal.
We can say that K_nearest and SVM do not perform well, while lin ress does. (It probably means that target value and features correlate linearly.) We will exclude K_nearest and SVM from our tests in the future.

Исходя из тестов выбираем заполнение пропущенных значений медианной из-за немного лучшего результата на кросс-валидации. Следующий шаг удаление выбросов. K_nearest и SVM плохо показали себя в тестах. В дальнейшем исключим их из наших тестов.

```python
labels = ['DecisionTree', 'LinRess', 'Ridge', 'RandomForest']
clf = [dtr, linres, ridge, rf]
```

```python
# applying standarization and removing outliers. assuming that the data have normal
# destribution and therefore the outliers lie 3 standart deviations.
# применяем стандаризацию и удаляем выбросы. полагаем, что данные имеют нормальное распределение
# и таким образом считаем, что выбросы лежита на расстоянии 3-ех стандартных отклонений.
from scipy import stats
df_train = df_median.copy()
df_train = df_train[(np.abs(stats.zscore(df_train)) < 3).all(axis=1)]
df_train = df_train.reset_index(drop=True)
X_train = df_train.drop(columns=['Energ_Kcal'])
y_train = df_train['Energ_Kcal']

score_train = []
for model in clf:
    score_train.append(-1*cross_val_score(model, X_train, y_train, cv=cv, scoring='neg_mean_absolute_error').mean())
print(score_train)
sns.barplot(x=labels, y=score_train)
```

In our next step we build and adjust our models based on all available features. Then we choose the best model based on model performance, learning time and model complexity. We also exclude DecisionTree from our tests.

Следующим шагом мы строим и настраиваем модели на всех доступных признаках. Затем мы выбираем лучшую модель исходя из эффективности, время на обучения и сложности модели. Также исключаем Дерево Решений из наших тестов.

So we end up with two choices: LineRess and Rigid. However the differences are so minar that we simpy choose the simplest model and thats the LinearRegression. Classification

Классификация

```python
from sklearn.feature_extraction.text import CountVectorizer
text_encoder = CountVectorizer(min_df=1)
```

```python
# form dataframe with encoded words
# сфопмиповываем датафрейм с закодированными словами
desc = df.copy().iloc[:,0]
desc = desc.replace('[^a-zA-Z]', ' ', regex = True).map(lambda x: x.lower()) # abra-cadabra and numbers removal in description column
desc = text_encoder.fit_transform(desc) # applying text encoder
features_name = text_encoder.get_feature_names() 
desc = pd.DataFrame(desc.toarray(), columns=features_name) # form dataframe of encoded features

desc['class'] = np.nan
```

```python
# choose words related to specific classes
# выбираем слова которые относятся к определенному классу

# bakery
# выпечка
temp_bakery = desc[(desc['bread'] !=0) | (desc['cake'] !=0) \
                   | (desc['wheat'] !=0) | (desc['bake'] !=0)\
                   | (desc['dough'] !=0) | (desc['flour'] !=0)\
                   | (desc['pizza'] !=0) | (desc['egg'] !=0)\
                   | (desc['biscuit'] !=0) | (desc['cookies'] !=0)\
                   | (desc['baked'] !=0)].index
# liquid
# напитки
temp_liquid = desc[(desc['water'] !=0) | (desc['alcoholic'] !=0) \
                  | (desc['beer'] !=0) | (desc['juice'] !=0)\
                  | (desc['beverage'] !=0) | (desc['tea'] !=0)\
                  | (desc['soda'] !=0) | (desc['smoothie'] !=0)\
                  | (desc['coffee'] !=0) | (desc['wine'] !=0)\
                  | (desc['cola'] !=0) | (desc['lemonade'] !=0)\
                  | (desc['drinks'] !=0)].index
# dairy products
# молочные продукты
temp_milkyway = desc[(desc['milk'] !=0) | (desc['yogurt'] !=0) \
                    | (desc['dairy'] !=0) | (desc['cheese'] !=0)\
                    | (desc['sour'] !=0)].index
# meat products
# мясная продукция
temp_savage = desc[(desc['beef'] !=0) | (desc['chicken'] !=0) \
                  | (desc['ham'] !=0) | (desc['pork'] !=0)\
                  | (desc['lamb'] !=0)].index
# vegetables and fruits
# овощи и фрукты
temp_vegan = desc[(desc['zucchini'] !=0) | (desc['corn'] !=0) \
                 | (desc['potato'] !=0) | (desc['cucumber'] !=0)\
                 | (desc['beans'] !=0) | (desc['banana'] !=0) \
                 | (desc['bananas'] !=0) | (desc['blueberry'] !=0)\
                 | (desc['vegetables'] !=0) | (desc['garlic'] !=0)\
                 | (desc['pepper'] !=0) | (desc['vegetarian'] !=0)].index

# check how many products we classified
# смотрим сколько продуктов мы классифицировали
temp_bakery.shape, temp_liquid.shape, temp_milkyway.shape, temp_savage.shape, temp_vegan.shape
```

```python
# assigning each product type specific number
# присваиваем каждому продукту специальный номер
desc.loc[temp_bakery, 'class'] = 1
desc.loc[temp_liquid, 'class'] = 2
desc.loc[temp_milkyway, 'class'] = 3
desc.loc[temp_savage, 'class'] = 4
desc.loc[temp_vegan, 'class'] = 5

# devide dataset to two ones. one with assigned classes other one without.
# делим датасет на два датасета. один с присвоенными классами, другой без
class_train_df = desc[desc['class'].notnull()] # returns class with no nulls
class_predict_df = desc[desc['class'].isnull()] # returns class with no nulls
class_predict_df = class_predict_df.drop(columns=['class'])
```

```python
from sklearn.semi_supervised import LabelPropagation
labler = LabelPropagation()
```

```python
# train semi-supervised LabelPropogation model to predict missing classes
# обучаем модел частичного обучения LabelPropogation для того, чтобы спрогнозировать классы пропущенных значений
class_X = class_train_df.iloc[:,0:-1]
class_y = class_train_df.iloc[:,-1]
# fitting the data
# вствляем данные
labler.fit(class_X, class_y)
# making predictions
# делаем предсказание
class_prediction = labler.predict(class_predict_df)
# concatenate two dataframes together
# сцепляем два датафрейма вместе
class_predict_df['class'] = class_prediction
train_classifier = pd.concat([class_train_df, class_predict_df]).reset_index(drop=True) # Now I am complete! 
```

```python
# The problem is in test set we can fing words that are excluded from train set and vice versa. Unlucky SKlearn
# has no idea how to work with missing values. So its up to us to choose features that are both available in train and test set.
# Проблема в том, что в тестовой выборке могут не находиться слова, которые есть в обучающей выборке и наоборот. К сожалению,
# Sklear не умеет работать с пропущенными данными. Поэтому это наша задача выбрать фичи, которые бы были и в обучающей
# и в тестовой выборке
test_class = test.copy().iloc[:,0]
test_class = test_class.replace('[^a-zA-Z]', ' ', regex = True).map(lambda x: x.lower()) # abra-cadabra and numbers removal
text_encoder = CountVectorizer(min_df=2)
test_class = text_encoder.fit_transform(test_class) 
features_name = text_encoder.get_feature_names() 
test_class = pd.DataFrame(test_class.toarray(), columns=features_name)

available_features = test_class.columns
```

```python
# we end up with followind sparse dataset. Now we can train Classifier to predict product class, based on product description
# У нас остается следующий разряженный датасет. Теперь можно обучить классификатор, который будет предсказывать класс продукта
# основываясь на его описании
zag_zag =  train_classifier.loc[:,available_features]
zag_zag['class'] = train_classifier['class']
zag_zag = zag_zag.fillna(0)
```

```python
# importing classifiers that can handle large sparse dataset
# импортируем классификатор, который может справиться с большим разреженным датасетом
from sklearn.linear_model import LogisticRegression
logRes = LogisticRegression(penalty='l2', random_state=0, solver='sag')
from sklearn.svm import SVC
# svm_gauss = SVC(random_state=0) # takes too long to compute
svm_linear = SVC(kernel='linear', random_state=0)
from sklearn.naive_bayes import GaussianNB
bayes = GaussianNB()

labels = ['LogRess', 'Linear_SVM', 'Bayes']
clf = [logRes, svm_linear, bayes]
```

```python
cv = KFold(5, True, 0)
X_zag_zag = zag_zag.drop(columns=['class'])
y_zag_zag = zag_zag['class']

score = []
for model in clf:
    score.append(cross_val_score(model, X_zag_zag, y_zag_zag, cv=cv, scoring='r2').mean())
print(score)
sns.barplot(x=labels, y=score)
```

```python
# Linear SVM wins with minor lead. Now lets build the model and predict classes for our test set
# Линейный SVM побеждает с незначительным отрывом. Теперь построим модель и прдскажем классы для тестовой выборки
svm_linear.fit(X_zag_zag, y_zag_zag)
test_classes= svm_linear.predict(test_class)
test_classes = pd.Series(test_classes.astype(int))
```
